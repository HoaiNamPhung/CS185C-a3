```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from operator import itemgetter
```

```python
# Load .tsv data.
customer_file = "CS185C-a3/maxcuststats.tsv"
product_file = "CS185C-a3/maxprodstats.tsv"
colnames = ["Stars", "Helpfulness"]
customers = np.loadtxt(customer_file, delimiter='\t')
sorted_cust = pd.DataFrame(sorted(customers, key=itemgetter(1)), columns=colnames)
products =  np.loadtxt(product_file, delimiter='\t')
sorted_prod = pd.DataFrame(sorted(products, key=itemgetter(1)), columns=colnames)
print(sorted_cust)
print(sorted_prod)
```

```python
# Plot the customer data.
# 1st col. = star rating, 2nd col. = helpfulness votes
plt.figure(figsize=(20,5))
plt.ylim(0,50)
plt.plot(sorted_cust.loc[:,"Helpfulness"], label="Helpfulness Votes", color="darkgreen")
plt.plot(sorted_cust.loc[:,"Stars"], label="Star Rating", color="orange")
plt.title("Customer Helpfulness Votes (Green) vs. Star Rating (Orange)")
plt.show
```

```python
# Plot the product data.
# 1st col. = star rating, 2nd col. = helpfulness votes
plt.figure(figsize=(20,5))
plt.ylim(0,50)
plt.plot(sorted_prod.loc[:,"Helpfulness"], label="Helpfulness Votes", color="darkgreen")
plt.plot(sorted_cust.loc[:,"Stars"], label="Star Rating", color="orange")
plt.title("Product Helpfulness Votes (Green) vs. Star Rating (Orange)")
plt.show
```
